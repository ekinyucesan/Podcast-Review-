```python
columns_to_check = ['fish_catches', 'fleet_size', 'fishing_patents', 'fisheries_imports']

# Group by country and sum the selected columns
total_values_by_country = df.groupby('iso3')[columns_to_check].sum()

print("Country with maximum total value for each specified column across all years:")
for col in columns_to_check:
    max_country = total_values_by_country[col].idxmax()
    max_value = total_values_by_country[col].max()
    print(f"- {col}: {max_country} (Total: {max_value:,.2f})")
```

```python
selected_columns = [
    'aquaculture_employment',
    'total_aquaculture_tonnes',
    'fao_fisheries_member',
    'fisheries_exports',
    'fleet_size',
    'fish_catches'
]

# Create a DataFrame with only the selected columns
df_selected = df[selected_columns]

# Calculate the correlation matrix for the selected columns
correlation_matrix_selected = df_selected.corr()

# Create a heatmap for the selected columns
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_selected, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Selected Features')
plt.show()
```

```python
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Create a heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of All Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(20, 12)) # Increased figure size
sns.lineplot(data=df, x='year', y='gdp', hue='iso3', legend=False) # Remove default legend
plt.title('GDP Comparison Across Countries Over Time')
plt.xlabel('Year')
plt.ylabel('GDP')

# Annotate each line with the country's iso3 code at the last data point
for iso3_code in df['iso3'].unique():
    country_data = df[df['iso3'] == iso3_code].sort_values(by='year')
    last_year = country_data['year'].iloc[-1]
    last_gdp = country_data['gdp'].iloc[-1]
    plt.text(last_year, last_gdp, iso3_code, fontsize=9, verticalalignment='center')

plt.grid(True)
plt.show()
```
